```python
import numpy as np
import pandas as pd

df_covid = pd.read_csv('002_2.COVID-19.csv')
df_subway = pd.read_csv('002_1.DAILY_SUBWAY_INFO.csv')

df_subway['date']= df_subway['date'].astype('str')
df_subway['date'] = pd.to_datetime(df_subway['date'])


df_subway = df_subway.drop(['on_num', 'off_num'], axis=1)
df_subway.columns = ['date', 'subway']
df_subway
```

```python
df = df_subway.copy()
df['COVID-19']=[0]*943
df
```

```python
df.set_index('date', drop=False, inplace = True)
df
```

```python

import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(15,10))

plt.plot('date',

         'Subway',

         data=df, 

         linestyle='none', 

         marker='o', 

         markersize=10,

         color='blue', 

         alpha=0.5)

plt.title("Scatter Plot of 'df' by matplotlib", fontsize=20)

plt.xlabel('date', fontsize=14)

plt.ylabel('Subway', fontsize=14)

plt.show()


```

```python

plt.figure(figsize=(15,10))

plt.plot('date',

         'Subway',

         data=df, 

         linestyle='solid', 

         marker='o', 

         markersize=0,

         color='blue', 

         alpha=0.5)

plt.title("Scatter Plot of 'df' by matplotlib", fontsize=20)

plt.xlabel('date', fontsize=14)

plt.ylabel('Subway', fontsize=14)

plt.show()


```

```python
df_week = pd.DataFrame()

df_week['Subway'] = df.Subway.resample('7D').sum()
df_week['COVID-19'] = df['COVID-19'].resample('7D').sum()



df_week

```

```python
df_week.insert(0, 'date', df_week.index)
df_week
```

```python
plt.figure(figsize=(15,10))

plt.plot('date',

         'Subway',

         data=df_week, 

         linestyle='solid', 

         marker='o', 

         markersize=0,

         color='blue', 

         alpha=0.5)

plt.title("Scatter Plot of 'df_week' by matplotlib", fontsize=20)

plt.xlabel('date', fontsize=14)

plt.ylabel('Subway', fontsize=14)


plt.show()


```

```python
plt.figure(figsize=(15,10))

plt.plot('date',

         'COVID-19',

         data=df_week, 

         linestyle='solid', 

         marker='o', 

         markersize=0,

         color='blue', 

         alpha=0.5)

plt.title("Scatter Plot of 'df_week' by matplotlib", fontsize=20)

plt.xlabel('date', fontsize=14)

plt.ylabel('COVID-19', fontsize=14)


plt.show()


```

```python
fig, ax1 = plt.subplots()

fig.set_figwidth(15)
fig.set_figheight(10)

plt.plot('date',

         'COVID-19',
         
         data=df_week, 

         linestyle='solid', 

         marker='o', 

         markersize=0,

         color='blue', 

         alpha=0.5)

ax2 = ax1.twinx()

plt.plot('date',

         'Subway',
         
         data=df_week, 

         linestyle='solid', 

         marker='o', 

         markersize=0,

         color='red', 

         alpha=0.5)

plt.title("Scatter Plot of 'df_week' by matplotlib", fontsize=20)

plt.xlabel('date', fontsize=14)

plt.ylabel('COVID-19', fontsize=14)


plt.show()

# Subway 범주 (y축) 빠짐
```
