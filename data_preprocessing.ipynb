```python

import numpy as np
import pandas as pd


df_covid = pd.read_csv('서울시 코로나19 확진자 현황.csv')
df_subway = pd.read_csv('001_1.일별_승하차수_20190101_20210731.csv')

df_subway['date']= df_subway['date'].astype('str')
df_subway['date'] = pd.to_datetime(df_subway['date'])
df_subway = df_subway.drop(['on_num', 'off_num'], axis=1)
df_subway.columns = ['date', 'Subway']
df_subway
```

```python
df = df_subway.copy()
df['COVID-19']=[0]*943
df
```

```python
df.set_index('date', drop=False, inplace = True)
df
```

```python
df_covid
```

```python
df_covid.rename(columns={df_covid.columns[1]:'date'}, inplace=True)
df_covid
```

```python
for i, row in df_covid.iterrows():
    if df['date'].isin([row['date']]).any():
        df.loc[row['date'], 'COVID-19'] += 1

df
```
